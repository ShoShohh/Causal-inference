```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set seed for reproducibility
np.random.seed(1)

# Edit the sample size
size= 5000
```

```python
# Structural Causal Model 1
def structural_causal_model1(error1, error2):
    x1 = error1
    x2 = 0.8 * x1 + error2
    return(x1, x2)

# Structural Causal Model 2
def structural_causal_model2(error1, error2):
    x2 = error2
    x1 = 0.8 * x2 + error1
    return(x1, x2)
```

```python
# Function generates random numbers following a Gaussian distribution
def generate_normal_numbers(size, loc, std):
    return np.random.normal(loc, std, size)

# Generate error variables
error1 = generate_normal_numbers(size, 0, 1)
error2 = generate_normal_numbers(size, 0, 1)

# Plot the data points
plt.scatter(structural_causal_model1(error1, error2)[0], structural_causal_model1(error1, error2)[1])
plt.scatter(structural_causal_model2(error1, error2)[0], structural_causal_model2(error1, error2)[1])

# Add labels and title to the plot
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Scatter Plot with Error Variables following a Gaussian distribution')

# Display the plot
plt.show()
```
